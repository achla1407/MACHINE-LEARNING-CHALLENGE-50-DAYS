# **WEB SCRAPPING FOR DATA COLLECTION**


```python
import pandas as pd
import requests
from bs4 import BeautifulSoup#library for web scraping
url ='https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
page= requests.get(url)
soup= BeautifulSoup(page.text,'html')
soup
```

```python
#requests.get('https://www.ambitionbox.com/list-of-companies?campaign=1').text
```

```python
#finding table or elements through class
soup.find('table',class_='wikitable sortable')
```

```python
 table =soup.find_all('table')[1]#table mein bhi kaunsi table yeh pata laga lena ke liye indexing kr sakte hai
 table
```

```python
world_title =table.find_all('th')
world_title
```

```python
world_table_title = [title.text.strip() for title in world_title]
world_table_title
```

```python
import pandas as pd
df=pd.DataFrame(columns= world_table_title)
df
```

```python
column_data =table.find_all('tr')
column_data
```

```python
for row in column_data[1:]:
  row_data=row.find_all('td')
  individual_row_data=[data.text.strip() for data in row_data]
  #print(individual_row_data)
  length=len(df)
  df.loc[length]=individual_row_data
```

```python
df
```

```python
df.to_csv('world_table.csv',index=False)#to convert the data into csv
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(response.text,'lxml')
soup.prettify()
```

```python
soup.find_all('h1')[0].text.strip()
#resultant jo bhi hoga voh ek list hoga so you know how to work with list and what kind all operation will you be using , text extract karne ke liye .text use karo
```

```python
company_name = soup.find_all('div',class_='companyCardWrapper__companyDetails')
company_name
```

```python

def get_company_name(soup):
      company_name = soup.find_all('div',class_='companyCardWrapper__companyDetails')
      company_names = []
      for i in company_name:
            company_names.append(i.text.strip())
      return company_names
company_list_details = get_company_name(soup)
company_list_details
```

```python
def complete_details():
        for name in company_list_details:
           print(name.split('\n'))
complete_details()
```

```python
cleaned_data = [s.replace("\t", "") for s in company_list_details]
final_data =[d.replace("\n","") for d in cleaned_data]
final_data
import pandas as pd
df=pd.DataFrame(columns=final_data)
df
```
