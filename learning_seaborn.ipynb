# CATEGORIALPLOT
--STRIPPLOT

```python
import seaborn as sns
import matplotlib.pyplot as plt
tip = sns.load_dataset('tips')
sns.catplot(x="day", y="tip", data=tip, hue='day',jitter =1 ,kind="strip")#SYNTAX
#stripplot
```

**JITTER**

 This is particularly useful in scatter plots where many data points overlap, making it difficult to see the individual points clearly.For example, in Matplotlib, you can use the jitter technique to spread out the points slightly to avoid overlap.

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 35]

# Create a scatter plot with alpha set to 0.5-1
plt.scatter(x, y, alpha=1)

plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Scatter plot with transparency')

plt.show()
```

A swarm plot, on the other hand, is a variation of the strip plot. It adjusts the position of data points along the categorical axis to avoid overlap, giving a clearer view of the distribution.

 **--SWARMPLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt
tip = sns.load_dataset('tips')
sns.catplot(x="day", y="tip", data=tip, hue='day' ,kind="swarm")#SYNTAX categorialplot sns.swarmplot gives the same result
```

# HEATMAP

```python
import seaborn as sns
flights = sns.load_dataset('flights')
flights
```


**values:** The data column you want to summarize (e.g., Values).

**index:** The rows you want to group by (e.g., Category).

**columns**: The columns you want to group by (e.g., Region).

**aggfunc:** The aggregation function to apply (e.g., sum, mean, etc.).

**fill_value:** The value to use for filling missing values (optional).

```python
flights_pivot = flights.pivot_table(values='passengers', index='year', columns='month', aggfunc='sum')
flights_pivot
```

```python
sns.heatmap(flights_pivot, annot=True, cmap='YlGnBu', fmt='d', linewidths=.5,cbar=False)
```
